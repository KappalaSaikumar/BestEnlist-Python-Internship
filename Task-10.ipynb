```python
# Exercise-2

def func2(txt):
    split=re.split("\s",txt)
    list1=[]
    for word in split:
        if re.search("ab",word):
            list1.append(word)
    return list1
```

```python
func2("iabcd abc abcd abcde jhvcwvVAba")
```

```python
# Exercise -3

def func3(txt):
    print(re.search("[0-9]*$",txt).group()," is the value found")
    print(re.search("[0-9]*$",txt).span()," is the position of the value found")
```

```python
func3("dhbcsdubcy356")
```

```python
func3("jahbcja3456")
```

```python
func3("hjbc463travgvc")
```
