```python
class CoffeeMaker(object):
    
    Report={"Water":1000,
        "Milk":1000,
        "Coffee":100,
        "Money":20}
    
    def __init__(self):
        self.start()
    
    def start(self):
        self.name=input("What would you like : ")
        if self.name=="report":
            print(self.report())
        elif self.name=="espresso":
            espresso.__init__(self)
        elif self.name=="latte":
            latte.__init__(self)
        elif self.name=="cappuccino":
            cappuccino.__init__(self)
        else:
            print("Invalid")
            self.start()
            
    def __del__(self):
        print("Machine turned off")
         
    def report(self):
        print("Water: ",self.Report["Water"],"ml \nMilk: ",self.Report["Milk"],"ml \nCoffee: ",self.Report["Coffee"],"g \nMoney: $",self.Report["Money"])
        self.start()
    
class espresso(CoffeeMaker):
    def __init__(self):
        self.__emoney=float(input("Enter Money : "))
        if(self.Report["Water"]>=250 and self.Report["Milk"]>=200 and self.Report["Coffee"]>=30 and self.__emoney>=float(3)):
            self.Report["Water"]-=250
            self.Report["Milk"]-=200
            self.Report["Coffee"]-=30
            self.Report["Money"]+=3
            if(self.__emoney>3):
                print("Here's your change : ",str(self.__emoney-3))
            print("Here's your espresso")
            self.start()
        elif self.__emoney<3:
            print("Not enough money. Enjoy!")
            self.start()
        else:
            print("There are no enough resources .. Sorry !")
        
class latte(CoffeeMaker):
    def __init__(self):
        self.__lmoney=float(input("Enter Money : "))
        if(self.Report["Water"]>=200 and self.Report["Milk"]>=150 and self.Report["Coffee"]>=24 and self.__lmoney>=2.5):
            self.Report["Water"]-=200
            self.Report["Milk"]-=150
            self.Report["Coffee"]-=24
            self.Report["Money"]+=2.5
            if(self.__lmoney>2.5):
                print("Here's your change : ",str(self.__lmoney-2.5))
            print("Here's your latte")
            self.start()
        elif self.__lmoney<2.5:
            print("Not enough money. Enjoy!")
            self.start()
        else:
            print("There are no enough resources .. Sorry !")
        
class cappuccino(CoffeeMaker):
    def __init__(self):
        self.__cmoney=float(input("Enter Money : "))
        if(self.Report["Water"]>=275 and self.Report["Milk"]>=160 and self.Report["Coffee"]>=33 and self.__cmoney>=float(5.0)):
            self.Report["Water"]-=275
            self.Report["Milk"]-=160
            self.Report["Coffee"]-=33
            self.Report["Money"]+=5
            if(self.__cmoney>5):
                print("Here's your change : ",str(self.__cmoney-5))
            print("Here's your cappuccino. Enjoy!")
            self.start()
        elif self.__cmoney<5:
            print("Not enough money")
            self.start()
        else:
            print("There are no enough resources .. Sorry !")
        
```
