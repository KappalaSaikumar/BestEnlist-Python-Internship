```python
# Exercise-2

def calculator(num1,num2,operator):
    if operator=="+":
        return num1+num2
    elif operator=="-":
        return num1-num2
    elif operator=="*":
        return num1*num2
    elif operator=="/":
        return num1/num2
    
```

```python
# Exercise-2

def func():
    try:
        num1 = int(input("Enter number 1 : "))
        num2=int(input("Enter number 2 : "))
        operator=str(input("Enter operator : "))
    except Exception as e:
        print(str(e), "Error occured")
    else:
        return calculator(num1,num2,operator)
```
