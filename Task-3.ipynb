```python
# Exercise-1 

def dict_add(dict1,dict2):
    dict_res=dict1.copy()
    dict_res.update(dict2)
    
    return dict_res
```

```python
# Exercise-1

cricketers={
    "virat":"India",
    "Abd":"SouthAfrica",
    "KlRahul":"India",
    "Bumrah":"India",
    "Marnus":"Aus",
    "Kane":"NZ"
}

dict_res=dict_add(capitals,cricketers)
print(dict_res)
```

```python
# Exercise-3

def merge_lists(list_keys,list_values):
    dict_res={}
    for i in range(min(len(list_keys),len(list_values))):
        dict_res[list_keys[i]]=list_values[i]
    
    # Or we can also use zip()
    dict_res=dict(zip(list_keys,list_values))
    return dict_res
```

```python
# Exercise-3

list_keys=list(cricketers.keys())
list_values=list(cricketers.values())
print(list_keys)
print(list_values)

print(merge_lists(list_keys,list_values))
```

```python
# Exercise-5

def remove_common(set1,set2):
    set_res=set1.copy()
    set_intersec=set1.intersection(set2)
    #or
    set_intersec=set1&set2
    for set_val in set_intersec:
        set_res.remove(set_val)
        
    return set_res
```

```python
RCB={"Virat","Abd","Yuzi","Saini","Devdutt"}
print(remove_common(India,RCB))
```
