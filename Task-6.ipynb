```python
# Exercise-1

def func1(list1):
    j=0
    for i in list1:
        list1[j]=i+2
        j=j+1
    return list1
```

```python
# Exercise-1

print(func1([1,2,3,4,5]))
```

```python
# Exercise-2

def func2(num):
    res=""
    for i in range(num,0,-1):
        for j in range(i,0,-1):
            res+=str(j)+" "
        res+="\n"
    return res
```

```python
print(func2(10))
```

```python
# Exercise-3

def func3(num):
    list2=[]
    list2.append(0)
    list2.append(1)
    for i in range(2,num):
        list2.append(list2[i-1]+list2[i-2])
    return list2
```

```python
print(func3(20))
```

```python
# Exercise-4

# Amstrong number : A number for which the sum of cubes of all it's digits is equal to the same number.

def func4(num):
    s=str(num)
    res=0
    for i in range(len(s)):
        res+=int(s[i])**3
    if res==num:
        print(str(num)," is an amstrong number")
    else:
        print(str(num)," is not an amstrong number")
```

```python
func4(153)
```

```python
# Exercise-7
# Considering some ideal case of 1 month=30 days and an year=360 days
def func7(days):
    years=int(days/360)
    if years>=1:
        months=int(int(days-years*360)/30)
        if months>=1:
            days=int(days-months*30-years*360)
            print(str(days)," days ",str(months)," months ",str(years)," years")
        else:
            print(str(int(days-years*360))," days ",str(months)," months ",str(years)," years")
    else:
        months=int(days/30)
        if months>=1:
            days=int(days-months*30)
            print(str(days)," days ",str(months)," months ",str(years)," years")
        else:
            print(str(days)," days ",str(months)," months ",str(years)," years")
```

```python
func7(59)
```

```python
# Exercise-9

def func9(num1,num2,operation):
    if(operation=="+"):
        print(str(num1)," + ",str(num2)," = ",str(num1+num2))
    elif(operation=="-"):
        print(str(num1)," - ",str(num2)," = ",str(num1-num2))
    elif(operation=="*"):
        print(str(num1)," * ",str(num2)," = ",str(num1*num2))
    elif(operation=="/"):
        print(str(num1)," / ",str(num2)," = ",str(num1/num2))
```

```python
func9(2,3,"*")
```
