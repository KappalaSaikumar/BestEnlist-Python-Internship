```python
# Exercise-1

def list_program(n,add,delete):
    list1=[]
    for i in range(n):
        list1.append(i)
    list1.append(add)
    del list1[delete]
    largest=max(list1)
    smallest=min(list1)
    
    return largest,smallest
```

```python
# Exercise-1

print(list_program(10,5,6))
```

```python
# Exercise-2

def reverse_tuple(tuple_1):
    rev_tuple=()
    n=len(tuple_1)
    for i in range(n):
        rev_tuple=rev_tuple+(tuple_1[n-1-i],)
    
    return rev_tuple
```

```python
# Exercise-2

tuple_1=(1,2,3,4)
print(reverse_tuple(tuple_1))
```
